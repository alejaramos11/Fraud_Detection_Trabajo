# Proyecto Final: Análisis Exploratorio de Transacciones Financieras en Línea

**Diccionario de Variables**
|Variable|Descripción|
|-|-|
|**`step`**|Unidad de tiempo en el mundo real. Hora y fecha en la que se realizó la transacción.|
|**`type`**|Tipo de transacción en línea: *CASH_IN*, *CASH_OUT*, *DEBIT*, *PAYMENT*, *TRANSFER*.|
|**`amount`**|Monto de la transacción.|
|**`nameOrig`**|Cliente que incia la transacción.|
|**`oldbalanceOrg`**|Saldo inicial del cliente antes de la transacción.|
|**`newbalanceOrg`**|Nuevo saldo del cliente después de la transacción.|
|**`nameDest`**|Cliente destinatario de la transacción.|
|**`oldbalanceDest`**|Saldo inicial del destinatario antes de la transacción.|
|**`newbalanceDest`**|Nuevo saldo del destinatario después de la transacción.|
|**`isFraud`**|Indicador de transacción: *1: Fraudulenta*.|
|**`isFlaggedFraud`**|Indicador de intentos de transacción ilegales marcados por un sistema de alerta.|
